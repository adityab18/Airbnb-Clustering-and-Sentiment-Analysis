## 🏠 Airbnb Dataset Feature Descriptions

### Identification & URLs

id – Unique identifier for each Airbnb listing.

listing_url – Web link to the Airbnb listing page.

name – Title of the listing shown on Airbnb.

summary – Short description provided by the host about the listing.

space – Information about the physical space (rooms, layout, etc.).

description – Full detailed description of the property.

neighborhood_overview – Overview of the surrounding area or locality.

notes – Additional notes or special instructions from the host.

transit – Information about public transport or commute options nearby.

access – Details on guest access (what areas are shared/private).

interaction – How and when the host interacts with guests.

house_rules – Rules and guidelines for staying at the property.

thumbnail_url – Link to the listing’s thumbnail image.

### 👤 Host Information

host_id – Unique ID assigned to the host.

host_url – Link to the host’s Airbnb profile.

host_name – Name of the host.

host_since – Date when the host joined Airbnb.

host_location – Location (city, state, country) of the host.

host_about – Description or bio written by the host.

host_response_time – Average response time of the host (e.g., within an hour, within a few hours).

host_response_rate – Percentage of messages the host responds to.

host_acceptance_rate – Percentage of booking requests accepted by the host.

host_is_superhost – Whether the host is a Superhost (t/f).

host_neighbourhood – The neighbourhood area where the host is located.

host_listings_count – Number of listings managed by the host in this dataset.

host_total_listings_count – Total number of listings the host manages overall.

host_has_profile_pic – Whether the host’s profile includes a picture (t/f).

host_identity_verified – Whether the host’s identity is verified (t/f).

### 📍 Location Information

street – Full street address (may be partially masked).

neighbourhood – Neighbourhood name provided by the host.

neighbourhood_cleansed – Standardized Airbnb neighbourhood label.

city – City where the property is located.

state – State abbreviation.

zipcode – Postal/ZIP code of the listing.

market – Broader market area (e.g., “San Diego”).

smart_location – City and country in a formatted way.

country_code – Two-letter country code (e.g., “US”).

country – Full country name.

latitude – Latitude coordinate.

longitude – Longitude coordinate.

is_location_exact – Whether the coordinates are exact or approximate.

### 🏡 Property Details

property_type – Type of property (e.g., House, Apartment).

room_type – Type of room offered (Entire home, Private room, Shared room).

accommodates – Maximum number of guests allowed.

bathrooms – Number of bathrooms.

bedrooms – Number of bedrooms.

beds – Total number of beds.

bed_type – Type of bed provided (e.g., Real Bed, Sofa Bed).

amenities – List of amenities available (WiFi, Kitchen, Parking, etc.).

square_feet – Area of the property in square feet (if provided).

### 💰 Pricing and Fees

nightly_price – Price per night.

price_per_stay – Total price for a stay (may include multiple nights).

security_deposit – Refundable security deposit charged by the host.

cleaning_fee – One-time cleaning fee.

guests_included – Number of guests included in base price.

extra_people – Additional charge per extra guest.

minimum_nights – Minimum number of nights required for booking.

maximum_nights – Maximum number of nights allowed per booking.

### ⭐ Reviews & Ratings

number_of_reviews – Total number of reviews the listing has received.

number_of_stays – (If available) Number of confirmed stays.

first_review – Date of the first review.

last_review – Date of the most recent review.

review_scores_rating – Overall average rating (0–100).

review_scores_accuracy – Rating for listing accuracy.

review_scores_cleanliness – Rating for cleanliness.

review_scores_checkin – Rating for check-in process.

review_scores_communication – Rating for host communication.

review_scores_location – Rating for location.

review_scores_value – Rating for value for money.

### ⚖️ Booking and Policies

requires_license – Whether a license is required to operate the listing.

instant_bookable – Whether guests can instantly book without approval.

is_business_travel_ready – Whether the listing is suitable for business travelers.

cancellation_policy – Type of cancellation policy (flexible, strict, etc.).

require_guest_profile_picture – Whether guests need a profile picture to book.

require_guest_phone_verification – Whether phone verification is required.

## Removing unnecessary columns

```python
columns_to_keep = ['id', 'listing_url', 'latitude', 'longitude', 'zipcode',
                   'name', 'summary', 'space', 'description', 'neighborhood_overview', 
                   'notes', 'transit', 'access', 'interaction',
                   'house_rules', 'property_type', 'room_type', 'accommodates','bathrooms', 
                   'bedrooms', 'beds', 'bed_type','amenities',
                   'nightly_price', 'price_per_stay', 'security_deposit', 'cleaning_fee',
                   'guests_included', 'extra_people', 'minimum_nights', 'maximum_nights',
                   'host_id', 'host_url', 'host_since', 'host_about', 'host_response_time', 
                   'host_response_rate', 'host_is_superhost', 'neighbourhood_cleansed',
                   'host_total_listings_count', 'host_has_profile_pic', 'host_identity_verified',
                   'number_of_reviews', 'number_of_stays', 'review_scores_rating', 
                   'review_scores_accuracy', 'review_scores_cleanliness', 
                   'review_scores_checkin','review_scores_communication', 'review_scores_location',
                   'review_scores_value', 'requires_license', 'instant_bookable',
                   'is_business_travel_ready', 'cancellation_policy',
                   'require_guest_profile_picture', 'require_guest_phone_verification'
                   ]
```

```python
df = data[columns_to_keep]
df.head(4)
```

### Cleaning zipcodes

```python
df['zipcode'] = df.groupby('neighbourhood_cleansed')['zipcode'].ffill()
df['zipcode']
df['zipcode'] = df['zipcode'].fillna('92101')
```

```python
print(df['zipcode'].unique())
```

```python
df['zipcode'] = df['zipcode'].str.split('-').str[0]
df['zipcode'] = df['zipcode'].astype(int)
```

### Removing signs and puncuation from the data

```python
# define the money columns
money_cols = ['nightly_price', 'price_per_stay', 'security_deposit', 'cleaning_fee', 'extra_people']

# remove all $ and , 
df[money_cols] = df[money_cols].replace({'\$':'', ',':''}, regex = True)

# convert to type float
df[money_cols] = df[money_cols].astype(float)

# convert host_response_rate to float
df['host_response_rate'] = df['host_response_rate'].replace({'%':''}, regex = True)
df['host_response_rate'] = df['host_response_rate'].astype(float)

df.head(2)
```

```python
# unique values for host response time
display(df['host_response_time'].value_counts())

# how many are missing?
display(df['host_response_time'].isna().sum())
```

```python
df['host_response_time'] = df['host_response_time'].fillna('N/A')
df['host_response_rate'] = df['host_response_rate'].fillna('0')
```

## Encoding the data

```python
# identify columns containing only 't' and 'f'
binary_cols = [col for col in df.columns if df[col].isin(['t', 'f']).all()]

# convert to 1 and 0
df[binary_cols] = df[binary_cols].replace({'t': 1, 'f': 0})

# ensure they are integers
df[binary_cols] = df[binary_cols].astype(int)
```

```python
def fill_missing(df):
    """ 
    Helper function that takes in columns from a dataframe and fills in the NaNs 
    with the median grouped by neigborhood
    """
    num_cols = ['review_scores_rating', 'review_scores_accuracy',
               'review_scores_cleanliness', 'review_scores_checkin',
               'review_scores_communication', 'review_scores_location',
               'review_scores_value', 'cleaning_fee', 'security_deposit',]
    
    for col in num_cols:
        df[col] = df[col].fillna(df.groupby('neighbourhood_cleansed')[col].transform('mean'))
    
    return df 
```
